### 2020년

```python
df1 = pd.read_csv('../data/화성시 성연령별유동인구_2020.csv')
df1
```

```python
df2 = pd.read_csv('../data/화성시 요일별 유동인구_2020.csv')
df2
```

```python
df3 = pd.read_csv('../data/화성시 시간대별유동인구_2020.csv')
df3
```

```python
# df1, df2, df3에서 위도, 경도가 일치하는지 확인
common_df1_df2 = pd.merge(df1, df2, on=['lat', 'lon', 'STD_YM'], how='inner')
# common_df1_df3 = pd.merge(df1[['lat', 'lon']], df3[['lat', 'lon']], on=['lat', 'lon'], how='inner')
# common_df2_df3 = pd.merge(df2[['lat', 'lon']], df3[['lat', 'lon']], on=['lat', 'lon'], how='inner')

# 결과 출력
print("df1과 df2의 겹치는 위도, 경도 개수:", len(common_df1_df2))
# print("df1과 df3의 겹치는 위도, 경도 개수:", len(common_df1_df3))
# print("df2와 df3의 겹치는 위도, 경도 개수:", len(common_df2_df3))
```

```python
common_df1_df2
```

```python
# df1, df2, df3에서 위도, 경도가 일치하는지 확인
common_df1_df2_df3 = pd.merge(common_df1_df2, df3, on=['lat', 'lon', 'STD_YM'], how='inner')

# 결과 출력
print("df1, df2, df3의 겹치는 위도, 경도 개수:", len(common_df1_df2_df3))
```

```python
common_df1_df2_df3
```

```python
common_df1_df2_df3.to_csv('../data/2020년 유동인구.csv', index=False)
```
