```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
from shapely.geometry import Point

import folium
from folium.plugins import MarkerCluster, MiniMap

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
df2 = pd.read_csv('../data/2021년 유동인구.csv')
df2
```

```python
df3 = pd.read_csv('../data/2022년 유동인구.csv')
df3
```

### 2021, 2022 합친 유동인구

```python
df_pop_all = pd.concat([df2,df3], axis=0)
df_pop_all # 2021 + 2022
```

### 2022 유동인구

```python
df_pop_2022 = df3.copy()
df_pop_2022 # 2022
```

```python
# 열 이름 필터링 (성연령별)
filtered_columns = [col for col in df_pop_all.columns if col.startswith("m_") or col.startswith("w_")]

# 선택한 열과 다른 주요 열 포함
selected_columns = ["STD_YM", "lon", "lat"] + filtered_columns
df_pop_all = df_pop_all[selected_columns]
df_pop_all
```

```python
# 열 이름 필터링 (성연령별)
filtered_columns = [col for col in df_pop_2022.columns if col.startswith("m_") or col.startswith("w_")]

# 선택한 열과 다른 주요 열 포함
selected_columns = ["STD_YM", "lon", "lat"] + filtered_columns
df_pop_2022 = df_pop_2022[selected_columns]
df_pop_2022
```

#### 전체 유동인구(2021+2022)

```python
# STD_YM, lat, lon을 제외한 열들끼리 각 행 합 계산 -> 전체 유동인구 수
df_pop_all['pop_sum'] = df_pop_all.iloc[:, 3:].sum(axis=1)

# 결과 출력 (lat, lon, pop_avg 열만 선택)
df_pop_all_1 = df_pop_all[['lat', 'lon', 'pop_sum']]
df_pop_all_1
```

```python
df_pop_all_2 = df_pop_all_1.groupby(['lat', 'lon']).mean().reset_index()
df_pop_all_2
```

```python
df_pop_all_2.info()
```

```python
import geopandas as gpd
df_map = gpd.read_file('../data/1-14.화성시_격자.geojson')
```

```python
gpd.GeoDataFrame(df_pop_all_2, geometry=gpd.GeoSeries(df_pop_all_2.apply(lambda row: Point(row['lon'], row['lat']), axis=1)), crs="EPSG:4326")
```

```python
df_pop_all_2
```

```python
#df_pop_all_2.to_csv('김현정_전체_유동인구_point.csv', index=False)
```

```python
df_map
```

```python
# 포인트 데이터 준비
# geometry 열이 이미 준비되어 있으므로 바로 GeoDataFrame으로 변환
gdf_points = gpd.GeoDataFrame(df_pop_all_2, geometry=df_pop_all_2['geometry'], crs="EPSG:4326")

# 격자 데이터 (df_map)도 GeoDataFrame 형식이어야 함
# 격자 데이터의 'geometry' 열이 다각형(Polygon) 형태여야 합니다.
gdf_grid = gpd.GeoDataFrame(df_map, geometry=df_map['geometry'], crs="EPSG:4326")
```

```python
# 공간 조인 실행: 포인트가 격자 내에 포함되는 경우 조인
result_all = gpd.sjoin(gdf_points, gdf_grid, how="inner", op="within")
result_all
```

#### 2022년 

```python
# STD_YM, lat, lon을 제외한 열들끼리 각 행 합 계산 -> 전체 유동인구 수
df_pop_2022['pop_sum'] = df_pop_2022.iloc[:, 3:].sum(axis=1)

# 결과 출력 (lat, lon, pop_avg 열만 선택)
df_pop_2022_1 = df_pop_2022[['lat', 'lon', 'pop_sum']]
df_pop_2022_1
```

```python
df_pop_2022_2 = df_pop_2022_1.groupby(['lat', 'lon']).mean().reset_index()
df_pop_2022_2
```

```python
df_pop_2022_2.info()
```

```python
gpd.GeoDataFrame(df_pop_2022_2, geometry=gpd.GeoSeries(df_pop_2022_2.apply(lambda row: Point(row['lon'], row['lat']), axis=1)), crs="EPSG:4326")
```

```python
# 포인트 데이터 준비
# geometry 열이 이미 준비되어 있으므로 바로 GeoDataFrame으로 변환
gdf_points = gpd.GeoDataFrame(df_pop_2022_2, geometry=df_pop_2022_2['geometry'], crs="EPSG:4326")

# 격자 데이터 (df_map)도 GeoDataFrame 형식이어야 함
# 격자 데이터의 'geometry' 열이 다각형(Polygon) 형태여야 합니다.
gdf_grid = gpd.GeoDataFrame(df_map, geometry=df_map['geometry'], crs="EPSG:4326")
```

```python
# 공간 조인 실행: 포인트가 격자 내에 포함되는 경우 조인
result_2022 = gpd.sjoin(gdf_points, gdf_grid, how="inner", op="within")
result_2022
```

### 2021 + 2022년 격자별 유동인구의 평균

```python
result_all_avg = (
    result_all.groupby("gid", as_index=False)
    .agg({
        "pop_sum": "mean",         # pop_avg 합계
        "lon": "mean",           # lon의 첫 번째 값
        "lat": "mean",           # lat의 첫 번째 값
    })
)
result_all_avg.rename(columns={'pop_sum' : 'pop_avg'}, inplace=True)
result_all_avg
```

```python
result_all_avg.to_csv('../data/화성시_2021년, 2022년 격자별 유동인구의 평균.csv', index=False)
```

### 2022년 격자별 유동인구의 평균

```python
result_2022_avg = (
    result_2022.groupby("gid", as_index=False)
    .agg({
        "pop_sum": "mean",         # pop_avg 합계
        "lon": "mean",           # lon의 첫 번째 값
        "lat": "mean",           # lat의 첫 번째 값
    })
)
result_2022_avg.rename(columns={'pop_sum' : 'pop_avg'}, inplace=True)
result_2022_avg
```

```python
result_2022_avg.to_csv('../data/2022년 화성시 격자별 유동인구의 평균.csv', index=False)
```
