```python
import numpy as np
import pandas as pd
```

```python
df_학교 = pd.read_csv('../data/전국초중등학교위치표준데이터.csv',encoding='cp949')
df_학교.head(3)
```

### 화성시

```python
df_학교_화성_포함 = df_학교[df_학교['소재지지번주소'].str.contains('화성', na=False)]
```

```python
df_학교_화성_포함 = df_학교_화성_포함.reset_index(drop=True)
df_학교_화성_포함
```

```python
df_학교_화성_포함.교육지원청명.unique()
```

```python
df_학교_화성 = df_학교_화성_포함[df_학교_화성_포함.교육지원청명 == '경기도화성오산교육지원청']
df_학교_화성.reset_index(inplace=True,drop=True)
```

```python
df_학교_화성[['학교명','학교급구분','설립형태','본교분교구분','소재지지번주소', '소재지도로명주소', '경도', '위도']].to_csv('../화성시학교.csv',index=False)
```

### 하남시

```python
df_학교_하남_포함 = df_학교[df_학교['소재지지번주소'].str.contains('하남', na=False)]
```

```python
df_학교_하남_포함 = df_학교_하남_포함.reset_index(drop=True)
df_학교_하남_포함
```

```python
df_학교_하남_포함.교육지원청명.unique()
```

```python
df_학교_하남 = df_학교_하남_포함[df_학교_하남_포함.교육지원청명 == '경기도광주하남교육지원청']
df_학교_하남.reset_index(inplace=True,drop=True)
```

```python
df_학교_하남[['학교명','학교급구분','설립형태','본교분교구분','소재지지번주소', '소재지도로명주소', '경도', '위도']].to_csv('../data/타겟/하남시학교.csv',index=False)
```
