```python
# 평균을 계산할 열 리스트
columns_to_average = ['mon_pop', 'tue_pop', 'wed_pop', 'thu_pop', 'fri_pop', 'sat_pop', 'sun_pop']

# lon과 lat으로 그룹화하여 각 열의 평균 계산
df3_means = df3.groupby(['lon', 'lat'])[columns_to_average].mean().reset_index()

# 결과 출력
df3_means
```

### 2020년 화성시 요일별 유동인구

```python
df3
```

```python
# STD_YM이 202001부터 202012 사이의 데이터만 추출
df3_2020 = df3[(df3['STD_YM'] >= 202001) & (df3['STD_YM'] <= 202012)]

# 결과 확인
df3_2020
```

```python
# STD_YM이 202001부터 202012 사이의 데이터만 추출
df3_2021 = df3[(df3['STD_YM'] >= 202101) & (df3['STD_YM'] <= 202112)]

# 결과 확인
df3_2021
```

```python
# STD_YM이 202001부터 202012 사이의 데이터만 추출
df3_2022 = df3[(df3['STD_YM'] >= 202201) & (df3['STD_YM'] <= 202212)]

# 결과 확인
df3_2022
```

```python
df3_2020.to_csv('../data/화성시 요일별 유동인구_2020.csv', index=False)
```

```python
df3_2021.to_csv('../data/화성시 요일별 유동인구_2021.csv', index=False)
```

```python
df3_2022.to_csv('../data/화성시 요일별 유동인구_2022.csv', index=False)
```
